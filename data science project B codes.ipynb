```python
import pandas as pd
import re
from collections import Counter


file_path = 'facebook评论表.xlsx' 
df = pd.read_excel(file_path)


comments = df['comment'].dropna()


ai_tools = [
   'chatgpt', 'gpt-4', 'bard', 'claude', 'jasper', 'copy.ai', 'writesonic', 'deepl write', 'rytr',
    'dalle', 'midjourney', 'stable diffusion', 'runway ml', 'artbreeder', 'deepart', 'dream by wombo', 'paintschainer',
    'soundraw', 'aiva', 'murf.ai', 'synthesia', 'descript', 'lumen5',
    'github copilot', 'amazon codewhisperer', 'tabnine', 'kite',
    'hugging face transformers', 'cohere', 'ai21 labs', 'jurassic-2', 
    'google dialogflow', 'microsoft luis',
    'tensorflow', 'pytorch', 'scikit-learn', 'xgboost', 'lightgbm', 'h2o.ai',
    'google translate', 'deepl translator', 'ibm watson text to speech', 'speechmatics', 'openai whisper',
    'perplexity', 'elicit', 'neeva ai', 'wolframalpha',
    'datarobot', 'rapidminer', 'alteryx', 'tableau ai','magic school',
    'notion ai', 'khanmigo', 'quizlet ai','kangaroo','canva','leonardo',  'grammarly', 'quillbot', 'scribesense', 'deepl write', 'wordtune', 'writelab',
    
    # 虚拟导师和个性化学习
    'khanmigo', 'quizlet ai', 'dreambox', 'carnegie learning', 'socratic by google', 
    'tutorme', 'knowji', 'smart sparrow',

    # 课程生成与评估
    'notion ai', 'edmentum', 'otus', 'gradescope', 'knewton', 'openstax tutor',

    # 语言学习
    'duolingo', 'rosetta stone', 'busuu', 'speakai', 'lingvist',

    # 协作与内容创作
    'explain everything', 'padlet', 'edpuzzle', 'nearpod', 'pear deck',

    # 教育数据分析
    'turnitin', 'wolframalpha', 'ecree', 'canvas lms'
]


def tokenize(text):
    
    text = re.sub(r'[^\w\s]', '', text).lower()
   
    return text.split()


tool_counts = Counter()

for comment in comments:
    tokens = tokenize(comment)
    for tool in ai_tools:
        if tool in ' '.join(tokens):  
            tool_counts[tool] += 1


top_10_ai_tools = tool_counts.most_common(10)

print("Top 10 AI tools:")
for tool, count in top_10_ai_tools:
    print(f"{tool}: {count}")
```
